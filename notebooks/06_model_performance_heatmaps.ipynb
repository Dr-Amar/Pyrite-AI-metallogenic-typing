```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Updated results dictionary with all classes and proper structure
results = {
    "RF": {
        "Standard": {
            "cm": np.array([[348, 3, 0, 19, 2, 1],
                           [25, 57, 0, 0, 0, 0],
                           [0, 0, 135, 12, 0, 0],
                           [7, 1, 1, 371, 0, 1],
                           [7, 3, 0, 0, 19, 0],
                           [5, 1, 0, 0, 0, 24]]),
            "report": {
                'Orogenic gold': {'precision': 0.89, 'recall': 0.93, 'f1-score': 0.91, 'support': 373},
                'Porphyry': {'precision': 0.88, 'recall': 0.70, 'f1-score': 0.78, 'support': 82},
                'SEDEX': {'precision': 0.99, 'recall': 0.92, 'f1-score': 0.95, 'support': 147},
                'Sedimentary': {'precision': 0.92, 'recall': 0.97, 'f1-score': 0.95, 'support': 381},
                'Skarn': {'precision': 0.90, 'recall': 0.66, 'f1-score': 0.76, 'support': 29},
                'VMS': {'precision': 0.92, 'recall': 0.80, 'f1-score': 0.86, 'support': 30},
                'accuracy': 0.9155,
                'macro avg': {'precision': 0.92, 'recall': 0.83, 'f1-score': 0.87, 'support': 1042},
                'weighted avg': {'precision': 0.92, 'recall': 0.92, 'f1-score': 0.91, 'support': 1042}
            }
        },
        "SMOTE": {
            "cm": np.array([[337, 4, 0, 22, 8, 2],
                           [17, 65, 0, 0, 0, 0],
                           [1, 0, 138, 8, 0, 0],
                           [9, 2, 3, 366, 0, 1],
                           [2, 2, 0, 0, 25, 0],
                           [1, 1, 0, 0, 0, 28]]),
            "report": {
                'Orogenic gold': {'precision': 0.92, 'recall': 0.90, 'f1-score': 0.91, 'support': 373},
                'Porphyry': {'precision': 0.88, 'recall': 0.79, 'f1-score': 0.83, 'support': 82},
                'SEDEX': {'precision': 0.98, 'recall': 0.94, 'f1-score': 0.96, 'support': 147},
                'Sedimentary': {'precision': 0.92, 'recall': 0.96, 'f1-score': 0.94, 'support': 381},
                'Skarn': {'precision': 0.76, 'recall': 0.86, 'f1-score': 0.81, 'support': 29},
                'VMS': {'precision': 0.90, 'recall': 0.93, 'f1-score': 0.92, 'support': 30},
                'accuracy': 0.9203,
                'macro avg': {'precision': 0.89, 'recall': 0.90, 'f1-score': 0.89, 'support': 1042},
                'weighted avg': {'precision': 0.92, 'recall': 0.92, 'f1-score': 0.92, 'support': 1042}
            }
        },
        "RUC": {
            "cm": np.array([[272, 28, 8, 30, 25, 10],
                           [9, 64, 0, 1, 5, 3],
                           [0, 0, 136, 11, 0, 0],
                           [26, 3, 22, 324, 3, 3],
                           [0, 0, 0, 0, 29, 0],
                           [0, 1, 0, 0, 1, 28]]),
            "report": {
                'Orogenic gold': {'precision': 0.89, 'recall': 0.73, 'f1-score': 0.80, 'support': 373},
                'Porphyry': {'precision': 0.67, 'recall': 0.78, 'f1-score': 0.72, 'support': 82},
                'SEDEX': {'precision': 0.82, 'recall': 0.93, 'f1-score': 0.87, 'support': 147},
                'Sedimentary': {'precision': 0.89, 'recall': 0.85, 'f1-score': 0.87, 'support': 381},
                'Skarn': {'precision': 0.46, 'recall': 1.00, 'f1-score': 0.63, 'support': 29},
                'VMS': {'precision': 0.64, 'recall': 0.93, 'f1-score': 0.76, 'support': 30},
                'accuracy': 0.8186,
                'macro avg': {'precision': 0.73, 'recall': 0.87, 'f1-score': 0.77, 'support': 1042},
                'weighted avg': {'precision': 0.84, 'recall': 0.82, 'f1-score': 0.82, 'support': 1042}
            }
        }
    },
    "SVM": {
        "Standard": {
            "cm": np.array([[348, 9, 0, 13, 2, 1],
                           [19, 59, 0, 0, 4, 0],
                           [3, 0, 140, 4, 0, 0],
                           [4, 1, 2, 374, 0, 0],
                           [6, 1, 0, 0, 22, 0],
                           [3, 0, 0, 0, 0, 27]]),
            "report": {
                'Orogenic gold': {'precision': 0.91, 'recall': 0.93, 'f1-score': 0.92, 'support': 373},
                'Porphyry': {'precision': 0.84, 'recall': 0.72, 'f1-score': 0.78, 'support': 82},
                'SEDEX': {'precision': 0.99, 'recall': 0.95, 'f1-score': 0.97, 'support': 147},
                'Sedimentary': {'precision': 0.96, 'recall': 0.98, 'f1-score': 0.97, 'support': 381},
                'Skarn': {'precision': 0.79, 'recall': 0.76, 'f1-score': 0.77, 'support': 29},
                'VMS': {'precision': 0.96, 'recall': 0.90, 'f1-score': 0.93, 'support': 30},
                'accuracy': 0.9309,
                'macro avg': {'precision': 0.91, 'recall': 0.87, 'f1-score': 0.89, 'support': 1042},
                'weighted avg': {'precision': 0.93, 'recall': 0.93, 'f1-score': 0.93, 'support': 1042}
            }
        },
        "SMOTE": {
            "cm": np.array([[355, 12, 1, 7, 3, 3],
                           [8, 373, 0, 0, 0, 0],
                           [2, 0, 378, 1, 0, 0],
                           [12, 2, 1, 366, 0, 0],
                           [1, 0, 0, 0, 380, 0],
                           [0, 0, 0, 0, 0, 381]]),
            "report": {
                'Orogenic gold': {'precision': 0.94, 'recall': 0.93, 'f1-score': 0.94, 'support': 381},
                'Porphyry': {'precision': 0.96, 'recall': 0.98, 'f1-score': 0.97, 'support': 381},
                'SEDEX': {'precision': 0.99, 'recall': 0.99, 'f1-score': 0.99, 'support': 381},
                'Sedimentary': {'precision': 0.98, 'recall': 0.96, 'f1-score': 0.97, 'support': 381},
                'Skarn': {'precision': 0.99, 'recall': 1.00, 'f1-score': 0.99, 'support': 381},
                'VMS': {'precision': 0.99, 'recall': 1.00, 'f1-score': 1.00, 'support': 381},
                'accuracy': 0.9768,
                'macro avg': {'precision': 0.98, 'recall': 0.98, 'f1-score': 0.98, 'support': 2286},
                'weighted avg': {'precision': 0.98, 'recall': 0.98, 'f1-score': 0.98, 'support': 2286}
            }
        },
        "RUC": {
            "cm": np.array([[18, 4, 1, 3, 1, 1],
                           [3, 22, 0, 0, 3, 0],
                           [0, 0, 25, 3, 1, 0],
                           [1, 0, 0, 27, 0, 0],
                           [0, 0, 1, 0, 27, 0],
                           [0, 1, 0, 0, 0, 28]]),
            "report": {
                'Orogenic gold': {'precision': 0.82, 'recall': 0.64, 'f1-score': 0.72, 'support': 28},
                'Porphyry': {'precision': 0.81, 'recall': 0.79, 'f1-score': 0.80, 'support': 28},
                'SEDEX': {'precision': 0.93, 'recall': 0.86, 'f1-score': 0.89, 'support': 29},
                'Sedimentary': {'precision': 0.82, 'recall': 0.96, 'f1-score': 0.89, 'support': 28},
                'Skarn': {'precision': 0.84, 'recall': 0.96, 'f1-score': 0.90, 'support': 28},
                'VMS': {'precision': 0.97, 'recall': 0.97, 'f1-score': 0.97, 'support': 29},
                'accuracy': 0.8547,
                'macro avg': {'precision': 0.86, 'recall': 0.86, 'f1-score': 0.86, 'support': 170},
                'weighted avg': {'precision': 0.87, 'recall': 0.86, 'f1-score': 0.86, 'support': 170}
            }
        }
    },
    "GB": {
        "Standard": {
            "cm": np.array([[346, 1, 2, 20, 3, 1],
                           [17, 65, 0, 0, 0, 0],
                           [1, 0, 135, 11, 0, 0],
                           [10, 1, 3, 367, 0, 0],
                           [8, 2, 0, 0, 18, 1],
                           [1, 1, 0, 0, 0, 28]]),
            "report": {
                'Orogenic gold': {'precision': 0.90, 'recall': 0.93, 'f1-score': 0.92, 'support': 373},
                'Porphyry': {'precision': 0.93, 'recall': 0.79, 'f1-score': 0.86, 'support': 82},
                'SEDEX': {'precision': 0.96, 'recall': 0.92, 'f1-score': 0.94, 'support': 147},
                'Sedimentary': {'precision': 0.92, 'recall': 0.96, 'f1-score': 0.94, 'support': 381},
                'Skarn': {'precision': 0.86, 'recall': 0.62, 'f1-score': 0.72, 'support': 29},
                'VMS': {'precision': 0.93, 'recall': 0.93, 'f1-score': 0.93, 'support': 30},
                'accuracy': 0.9203,
                'macro avg': {'precision': 0.92, 'recall': 0.86, 'f1-score': 0.88, 'support': 1042},
                'weighted avg': {'precision': 0.92, 'recall': 0.92, 'f1-score': 0.92, 'support': 1042}
            }
        },
        "SMOTE": {
            "cm": np.array([[354, 8, 0, 13, 3, 3],
                           [11, 362, 0, 1, 7, 0],
                           [0, 0, 374, 7, 0, 0],
                           [12, 0, 0, 368, 0, 1],
                           [2, 0, 0, 0, 379, 0],
                           [0, 0, 0, 1, 0, 380]]),
            "report": {
                'Orogenic gold': {'precision': 0.93, 'recall': 0.93, 'f1-score': 0.93, 'support': 381},
                'Porphyry': {'precision': 0.98, 'recall': 0.95, 'f1-score': 0.96, 'support': 381},
                'SEDEX': {'precision': 1.00, 'recall': 0.98, 'f1-score': 0.99, 'support': 381},
                'Sedimentary': {'precision': 0.94, 'recall': 0.97, 'f1-score': 0.95, 'support': 381},
                'Skarn': {'precision': 0.97, 'recall': 0.99, 'f1-score': 0.98, 'support': 381},
                'VMS': {'precision': 0.99, 'recall': 1.00, 'f1-score': 0.99, 'support': 381},
                'accuracy': 0.9698,
                'macro avg': {'precision': 0.97, 'recall': 0.97, 'f1-score': 0.97, 'support': 2286},
                'weighted avg': {'precision': 0.97, 'recall': 0.97, 'f1-score': 0.97, 'support': 2286}
            }
        },
        "RUC": {
            "cm": np.array([[19, 3, 1, 4, 1, 0],
                           [3, 24, 0, 0, 1, 0],
                           [0, 0, 27, 2, 0, 0],
                           [3, 0, 1, 22, 2, 0],
                           [3, 0, 1, 0, 24, 0],
                           [0, 0, 0, 0, 0, 29]]),
            "report": {
                'Orogenic gold': {'precision': 0.68, 'recall': 0.68, 'f1-score': 0.68, 'support': 28},
                'Porphyry': {'precision': 0.89, 'recall': 0.86, 'f1-score': 0.87, 'support': 28},
                'SEDEX': {'precision': 0.90, 'recall': 0.93, 'f1-score': 0.92, 'support': 29},
                'Sedimentary': {'precision': 0.79, 'recall': 0.79, 'f1-score': 0.79, 'support': 28},
                'Skarn': {'precision': 0.86, 'recall': 0.86, 'f1-score': 0.86, 'support': 28},
                'VMS': {'precision': 1.00, 'recall': 1.00, 'f1-score': 1.00, 'support': 29},
                'accuracy': 0.8529,
                'macro avg': {'precision': 0.85, 'recall': 0.85, 'f1-score': 0.85, 'support': 170},
                'weighted avg': {'precision': 0.85, 'recall': 0.85, 'f1-score': 0.85, 'support': 170}
            }
        }
    },
    "MLP": {
        "Standard": {
            "cm": np.array([[339, 10, 0, 17, 6, 1],
                           [18, 60, 0, 0, 4, 0],
                           [1, 0, 142, 4, 0, 0],
                           [8, 1, 1, 371, 0, 0],
                           [5, 2, 0, 0, 22, 0],
                           [1, 2, 0, 1, 0, 26]]),
            "report": {
                'Orogenic gold': {'precision': 0.91, 'recall': 0.91, 'f1-score': 0.91, 'support': 373},
                'Porphyry': {'precision': 0.80, 'recall': 0.73, 'f1-score': 0.76, 'support': 82},
                'SEDEX': {'precision': 0.99, 'recall': 0.97, 'f1-score': 0.98, 'support': 147},
                'Sedimentary': {'precision': 0.94, 'recall': 0.97, 'f1-score': 0.96, 'support': 381},
                'Skarn': {'precision': 0.69, 'recall': 0.76, 'f1-score': 0.72, 'support': 29},
                'VMS': {'precision': 0.96, 'recall': 0.87, 'f1-score': 0.91, 'support': 30},
                'accuracy': 0.9213,
                'macro avg': {'precision': 0.88, 'recall': 0.87, 'f1-score': 0.87, 'support': 1042},
                'weighted avg': {'precision': 0.92, 'recall': 0.92, 'f1-score': 0.92, 'support': 1042}
            }
        },
        "SMOTE": {
            "cm": np.array([[342, 15, 2, 13, 5, 4],
                           [2, 376, 0, 2, 1, 0],
                           [1, 0, 379, 1, 0, 0],
                           [10, 2, 1, 367, 1, 0],
                           [0, 0, 0, 0, 381, 0],
                           [0, 0, 0, 0, 0, 381]]),
            "report": {
                'Orogenic gold': {'precision': 0.96, 'recall': 0.90, 'f1-score': 0.93, 'support': 381},
                'Porphyry': {'precision': 0.96, 'recall': 0.99, 'f1-score': 0.97, 'support': 381},
                'SEDEX': {'precision': 0.99, 'recall': 0.99, 'f1-score': 0.99, 'support': 381},
                'Sedimentary': {'precision': 0.96, 'recall': 0.96, 'f1-score': 0.96, 'support': 381},
                'Skarn': {'precision': 0.98, 'recall': 1.00, 'f1-score': 0.99, 'support': 381},
                'VMS': {'precision': 0.99, 'recall': 1.00, 'f1-score': 0.99, 'support': 381},
                'accuracy': 0.9738,
                'macro avg': {'precision': 0.97, 'recall': 0.97, 'f1-score': 0.97, 'support': 2286},
                'weighted avg': {'precision': 0.97, 'recall': 0.97, 'f1-score': 0.97, 'support': 2286}
            }
        },
        "RUC": {
            "cm": np.array([[18, 3, 0, 5, 1, 1],
                           [4, 21, 0, 0, 3, 0],
                           [0, 0, 24, 5, 0, 0],
                           [0, 0, 0, 27, 0, 1],
                           [0, 0, 1, 0, 27, 0],
                           [0, 0, 0, 0, 0, 28]]),
            "report": {
                'Orogenic gold': {'precision': 0.82, 'recall': 0.64, 'f1-score': 0.72, 'support': 28},
                'Porphyry': {'precision': 0.88, 'recall': 0.75, 'f1-score': 0.81, 'support': 28},
                'SEDEX': {'precision': 0.96, 'recall': 0.83, 'f1-score': 0.89, 'support': 29},
                'Sedimentary': {'precision': 0.73, 'recall': 0.96, 'f1-score': 0.83, 'support': 28},
                'Skarn': {'precision': 0.87, 'recall': 0.96, 'f1-score': 0.92, 'support': 28},
                'VMS': {'precision': 0.94, 'recall': 0.97, 'f1-score': 0.97, 'support': 29},
                'accuracy': 0.8588,
                'macro avg': {'precision': 0.86, 'recall': 0.86, 'f1-score': 0.85, 'support': 170},
                'weighted avg': {'precision': 0.87, 'recall': 0.86, 'f1-score': 0.86, 'support': 170}
            }
        }
    }
}

# Models and Resampling Methods
model_names = ["RF", "SVM", "GB", "MLP"]
resampling_methods = ["Standard", "SMOTE", "RUC"]
class_names = ['Orogenic gold', 'Porphyry', 'SEDEX', 'Sedimentary', 'Skarn', 'VMS']

# --- Plotting Functions ---

def plot_confusion_matrices(method):
    fig, axes = plt.subplots(2, 2, figsize=(18, 15))
    fig.suptitle(f"{method} - Confusion Matrices", fontsize=18)

    for ax, model in zip(axes.flatten(), model_names):
        cm = results[model][method]["cm"]
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax, cbar=True,
                   xticklabels=class_names, yticklabels=class_names)
        ax.set_title(model)
        ax.set_xlabel("Predicted")
        ax.set_ylabel("Actual")
        plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
        plt.setp(ax.get_yticklabels(), rotation=0)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

def plot_classification_reports(method):
    fig, axes = plt.subplots(2, 2, figsize=(18, 15))
    fig.suptitle(f"{method} - Classification Reports", fontsize=18)

    for ax, model in zip(axes.flatten(), model_names):
        report = results[model][method]["report"]
        # Extract only the class metrics (excluding accuracy and averages)
        metrics = {}
        for class_name in class_names:
            metrics[class_name] = {
                'precision': report[class_name]['precision'],
                'recall': report[class_name]['recall'],
                'f1-score': report[class_name]['f1-score']
            }
        df = pd.DataFrame(metrics).T
        sns.heatmap(df, annot=True, fmt=".2f", cmap="Greens", cbar=True, ax=ax, vmin=0, vmax=1)
        ax.set_title(model)
        ax.set_xlabel("Metrics")
        ax.set_ylabel("Classes")
        plt.setp(ax.get_xticklabels(), rotation=0)
        plt.setp(ax.get_yticklabels(), rotation=0)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

# --- Generate all plots ---
for method in resampling_methods:
    plot_confusion_matrices(method)
    plot_classification_reports(method)
```
